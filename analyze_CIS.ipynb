# Zusammenfassung über die Dateien
* CIS-Daten.xlsx liefert für konsistente Daten nur die Spalten `ANKUNFT` und `ABFAHRT`, was wahrscheinlich der Soll-Zustand ist und fast konsistente Daten in `TATS_ANKUNFT` und `TATS_ABFAHRT` (152/2800 Zeilen nicht konsistent), was wahrscheinlich der Ist-Zustand ist
* Aufträge.xlsx und AufträgeOhneDisposoll.xlsx liefern kein Ankunft oder Abfahrt, sondern `Ladedatum` in vielen Spalten 
* die ganzen einzelnen hinterlegten xml Dateien für den März liefern zwar Zeitstempel, aber keine Information über Abfahrt und Ankunft -> genauso bei den Daten in der Schnittstelle
## Löschen der 152 Zeilen der Erfassungsfehler

```python
df_filtered = df_filtered[df_filtered.tats_ankunft <= df_filtered.tats_abfahrt]
print(df_filtered.shape)
```

```python
df_filtered = df_filtered.drop(columns=['ankunft_sollvon', 'ankunft_sollbis', 'ankunftdatum', 'startdatum', 'istzeitstatus'])
```

### Datentyp object für Ankunft und Abfahrt in dataetime umwandeln

```python
df_filtered.dtypes
```

```python
dt_cols = ['ankunft','tats_ankunft','abfahrt','tats_abfahrt']
for col in dt_cols:
    if col in df_filtered.columns:
        df_filtered[col] = pd.to_datetime(
            df_filtered[col]
            .astype(str)
            .str.replace("\xa0", " ", regex=False)
            .str.strip(),
            format="%d.%m.%Y %H:%M",
            errors="coerce"
        )
        # Also try date-only format
        mask = df_filtered[col].isna()
        if mask.any():
            df_filtered.loc[mask, col] = pd.to_datetime(df_filtered.loc[mask, col], errors='coerce')
df_filtered
                
```
