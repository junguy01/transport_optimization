## Analyse Fahrerverhalten

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Ist-Daten.csv', sep=',', encoding='utf-8')  # sep=',' falls Komma-getrennt

# Erste Übersicht
print(df.shape)
print(df.dtypes)
df.head()
```

```python
# Zeitstempel konvertieren
df['Abfahrt'] = pd.to_datetime(df['Abfahrt'])
df['Ankunft'] = pd.to_datetime(df['Ankunft'])

# Standzeit berechnen (nur wo beide vorhanden)
df['Standzeit_min'] = (
    df['Abfahrt'] - df['Ankunft']
).dt.total_seconds() / 60
```

```python
# Schritt 1: Originale Reihenfolge merken
df = df.reset_index(drop=True)
df['_order'] = df.index

# Schritt 2: Konsekutive Gruppe identifizieren
df['_new_block'] = (
    (df['TOURNR'] != df['TOURNR'].shift()) |
    (df['Position'] != df['Position'].shift())
).cumsum()

# Schritt 3: Innerhalb jedes Blocks zusammenführen
df_merged = (
    df.groupby(['TOURNR', 'Position', '_new_block'], as_index=False)
    .agg(
        Abfahrt=('Abfahrt', 'first'),
        Ankunft=('Ankunft', 'first'),
        Fahrername=('Fahrername', 'first'),
        KENNZ_CLEAN=('KENNZ_CLEAN', 'first'),
        Längengrad=('Längengrad', 'first'),
        Breitengrad=('Breitengrad', 'first'),
        _order=('_order', 'first'),  # kleinsten Index des Blocks merken
    )
)

# Schritt 4: Nach originalem Index sortieren, Hilfsspalten entfernen
df_merged = (
    df_merged
    .sort_values('_order')
    .drop(columns=['_new_block', '_order'])
    .reset_index(drop=True)
)

print(f"Zeilen vorher: {len(df)}")
print(f"Zeilen nachher: {len(df_merged)}")
df_merged.head(20)
```

```python
# Index sicherstellen
df_merged = df_merged.reset_index(drop=True)

# Koordinaten konvertieren
if df_merged['Breitengrad'].dtype == object:
    df_merged['Breitengrad'] = df_merged['Breitengrad'].astype(str).str.replace(',', '.').astype(float)
    df_merged['Längengrad'] = df_merged['Längengrad'].astype(str).str.replace(',', '.').astype(float)

RADIUS = 1000

# Haversine
lat1 = np.radians(df_merged['Breitengrad'])
lon1 = np.radians(df_merged['Längengrad'])
lat2 = np.radians(df_merged['Breitengrad'].shift())
lon2 = np.radians(df_merged['Längengrad'].shift())

dlat = lat1 - lat2
dlon = lon1 - lon2

a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
df_merged['_dist'] = 6371000 * 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

# NUR Vorgänger-Bedingung: eine hat Ankunft=NaN, direkt vorherige hat Abfahrt=NaN (oder umgekehrt)
merge_erlaubt = (
    (df_merged['Ankunft'].isna() & df_merged['Abfahrt'].shift().isna()) |
    (df_merged['Abfahrt'].isna() & df_merged['Ankunft'].shift().isna())
)

print("merge_erlaubt True:", merge_erlaubt.sum())  # sollte jetzt 204 sein

df_merged['_new_block'] = (
    (df_merged['TOURNR'] != df_merged['TOURNR'].shift()) |
    (df_merged['_dist'] > RADIUS) |
    ~merge_erlaubt
).cumsum()

df_merged['_order'] = df_merged.index

df_merged2 = (
    df_merged.groupby(['TOURNR', '_new_block'], as_index=False)
    .agg(
        Position=('Position', 'first'),
        Abfahrt=('Abfahrt', 'first'),
        Ankunft=('Ankunft', 'last'),
        Fahrername=('Fahrername', 'first'),
        KENNZ_CLEAN=('KENNZ_CLEAN', 'first'),
        Längengrad=('Längengrad', 'first'),
        Breitengrad=('Breitengrad', 'first'),
        _order=('_order', 'first'),
    )
)

df_merged2 = (
    df_merged2
    .sort_values('_order')
    .drop(columns=['_new_block', '_order', '_dist'], errors='ignore')
    .reset_index(drop=True)
)

print(f"Zeilen zu beginn: {len(df)}")
print(f"Zeilen nach Pass 1 (Position): {len(df_merged)}")
print(f"Zeilen nach Pass 2 (Geofencing): {len(df_merged2)}")
df_merged2.head(10)
```
